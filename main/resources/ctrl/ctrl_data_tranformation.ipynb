```python
import os
import numpy as np
import pandas as pd

# 원본 CSV 파일 읽기 (첫 번째 열을 인덱스로 사용)
ctrl_data = pd.read_csv('/home/user/workdir/CMAQ_Emulator/main/resources/ctrl/precursor_control_2013.csv', index_col=0)

# NOX, VOC, NH3, PM25가 포함된 모든 컬럼을 추출
filtering_columns = ctrl_data.columns[ctrl_data.columns.str.contains('NOX|VOC|NH3|PM25')]
transformed_data = ctrl_data[filtering_columns].copy()

# 각 컬럼의 접두어(예: 'A', 'B', ..., 'Q') 추출 
# (컬럼명 형식이 LETTER_NOX_ALL, LETTER_VOC_ALL, LETTER_NH3_ALL, LETTER_PM25_ALL 등이어야 함)
letters = sorted({col.split('_')[0] for col in transformed_data.columns if '_' in col})

# 각 접두어별로 NOX, VOC 관련 추가 interaction term 생성
for letter in letters:
    nox_col = f"{letter}_NOX_ALL"
    voc_col = f"{letter}_VOC_ALL"
    # 해당 그룹의 NOX와 VOC 컬럼이 존재할 때만 계산
    if nox_col in transformed_data.columns and voc_col in transformed_data.columns:
        # NOX²
        transformed_data[f"{letter}_NOX_SQUARE_ALL"] = transformed_data[nox_col] ** 2
        # VOC²
        transformed_data[f"{letter}_VOC_SQUARE_ALL"] = transformed_data[voc_col] ** 2
        # NOX * VOC (기존)
        transformed_data[f"{letter}_NOX_VOC_ALL"] = transformed_data[nox_col] * transformed_data[voc_col]
        # (NOX²) * VOC
        transformed_data[f"{letter}_NOX_SQUARE_VOC_ALL"] = (transformed_data[nox_col] ** 2) * transformed_data[voc_col]
        # NOX * (VOC²)
        transformed_data[f"{letter}_NOX_VOC_SQUARE_ALL"] = transformed_data[nox_col] * (transformed_data[voc_col] ** 2)

# 최종 컬럼 순서를 재정렬:
# 각 접두어별로 아래 순서로 컬럼을 나열:
# NOX_ALL, VOC_ALL, NOX_SQUARE_ALL, VOC_SQUARE_ALL, NOX_VOC_ALL, NOX_SQUARE_VOC_ALL, NOX_VOC_SQUARE_ALL, NH3_ALL, PM25_ALL
ordered_columns = []
for letter in letters:
    for suffix in ['NOX_ALL', 'VOC_ALL', 'NOX_SQUARE_ALL', 'VOC_SQUARE_ALL',
                   'NOX_VOC_ALL', 'NOX_SQUARE_VOC_ALL', 'NOX_VOC_SQUARE_ALL',
                   'NH3_ALL', 'PM25_ALL']:
        colname = f"{letter}_{suffix}"
        if colname in transformed_data.columns:
            ordered_columns.append(colname)

transformed_data = transformed_data[ordered_columns]

# 결과 CSV 파일 저장 (예: 4 input 특성+추가 interaction term으로 구성된 데이터셋)
output_file_path = '/home/user/workdir/CMAQ_Emulator/main/resources/ctrl/dataset_for_o3_interaction_v1.csv'
transformed_data.to_csv(output_file_path)

print(f"Filtered data saved to {output_file_path}")

# 최종 결과 확인
transformed_data
```
