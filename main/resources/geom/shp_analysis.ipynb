```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os

# -- 이 부분에 검증하고 싶은 Shapefile의 전체 경로를 입력하세요 --
# 예시: shp_path = '/home/user/workdir/CMAQ_Emulator/main/resources/geom/ctp_rvn.shp'
# 예시: shp_path = '/home/user/workdir/CMAQ_Emulator/main/resources/geom/TL_SCCO_SIG.shp'
# 예시: shp_path = '/home/user/workdir/CMAQ_Emulator/main/resources/geom/si_gun_gu/SKOR.shp'

shp_path = '/home/user/workdir/CMAQ_Emulator/main/resources/geom/TL_SCCO_SIG.shp' # <--- 여기를 바꿔가며 실행

# Shapefile 읽기 (SHX 복구 옵션 포함)
os.environ['SHAPE_RESTORE_SHX'] = 'YES'
try:
    gdf = gpd.read_file(shp_path, encoding='cp949')

    print(f"=================================================================")
    print(f"📄 파일 검증: {shp_path}")
    print(f"=================================================================")
    
    # 1. 총 폴리곤(지역) 개수 출력
    print(f"\n[정보 1] 총 지역(폴리곤) 개수: {len(gdf)}\n")
    
    # 2. 컬럼 정보 출력
    print("[정보 2] 컬럼 상세 정보:")
    gdf.info()
    
    # 3. 실제 데이터 샘플 출력
    print("\n[정보 3] 실제 데이터 (상위 5개):")
    # geometry 컬럼은 너무 길어서 제외하고 출력
    display(gdf.drop(columns='geometry').head())
    
    # 4. 지도 시각화
    print("\n[정보 4] 지도 시각화:")
    gdf.plot()
    plt.title(os.path.basename(shp_path))
    plt.show()

except Exception as e:
    print(f"파일을 읽는 중 에러가 발생했습니다: {e}")
```

```python
import pandas as pd
import geopandas as gpd

# 1. SHP 파일 로드 (전체 251개 지역)
sigungu_map_path = '/home/user/workdir/CMAQ_Emulator/main/resources/geom/TL_SCCO_SIG.shp'
sigungu_map = gpd.read_file(sigungu_map_path, encoding='cp949')

# 2. CSV 파일 로드 (분석 대상 201개 지역)
grid_alloc_path = '/home/user/workdir/CMAQ_Emulator/main/resources/geom/si_gun_gu_allocation.csv'
grid_alloc = pd.read_csv(grid_alloc_path)

# 3. 각 파일에 포함된 고유 시군구 이름(영문) 추출
# SIG_ENG_NM 컬럼의 결측치(NaN)는 제거
shp_sigungu_list = set(sigungu_map['SIG_ENG_NM'].dropna())
csv_sigungu_list = set(grid_alloc['SIG_ENG_NM'].dropna()) # CSV의 영문 컬럼명은 'SIG_ENG_NM'으로 가정

# 4. 차집합을 이용해 CSV에 누락된 지역 찾기
missing_sigungu = shp_sigungu_list - csv_sigungu_list

print(f"SHP 파일의 고유 시군구 개수: {len(shp_sigungu_list)}")
print(f"CSV 파일의 고유 시군구 개수: {len(csv_sigungu_list)}")
print(f"분석에서 제외된 시군구 개수: {len(missing_sigungu)}")
print("\n--- 분석에서 제외된 지역 (일부) ---")
print(sorted(list(missing_sigungu))[:10]) # 처음 10개만 출력
```
